## DATA ANALYSIS

```python
# Loading the data into a pandas dataframe
df = emp_data.copy()
df
```

##### Checking for distribution 

Since there no categorical values, distribution wil not be cheked.

However an exploratory data analysis will be made to find out the number of employees according to department, region, education, gender, recruitment channel and full time/part time workers.

#### DEPARTMENT

```python

# Employee distribution by department
sns.countplot(x='Department', data=df)
sns.set(rc={'figure.figsize':(12,13)})
plt.title('Employee distribution by department')
plt.show()
```

The sales and marketing department have the highesT number of employess 

#### REGION

```python
reg_counts = df.groupby('Region')['Employee ID'].count().nlargest(5).to_frame()
reg_counts
```

```python
reg_counts = df.groupby('Region')['Employee ID'].count().nlargest(5)
# Create a bar chart to find the region with the highest number of employees.
plt.bar(reg_counts.index, reg_counts.values)
plt.title('Employee Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Number of Employees')
plt.show()

```

According to this graph, most employees are from region_2.

#### EDUCATION

```python
edu_counts = df.groupby('Education')['Employee ID'].count().to_frame()
edu_counts
```

```python
# Create a bar chart to find the region with the highest number of employees.
plt.bar(edu_counts.index, edu_counts.values)
plt.title('Employee Distribution by Education')
plt.xlabel('Education')
plt.ylabel('Number of Employees')
plt.show()
```

```python
# Employee distribution by education
sns.countplot(x='Education', data=df)
plt.title('Employee distribution by education')
```

#### GENDER

```python
#Employee distribution by gender
sns.countplot(x='Gender', data=df)
plt.title('Employee distribution by gender')
```

#### RECUITMENT CHANNEL
